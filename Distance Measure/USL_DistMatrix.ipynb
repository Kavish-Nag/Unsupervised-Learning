```python
from sklearn.metrics import DistanceMetric
dist=DistanceMetric.get_metric('euclidean') #uses euclidean form to calculate distance
X=[[1,2],[3,4],[5,6]]
Y=[[7,8],[9,10]]
dist.pairwise(X,Y)
```

```python
from sklearn.metrics import DistanceMetric
dist=DistanceMetric.get_metric('manhattan') #uses manhattan form to calculate distance
X=[[1,2],[3,4],[5,6]]
Y=[[7,8],[9,10]]
dist.pairwise(X,Y)
```

```python
from sklearn.metrics import DistanceMetric
dist=DistanceMetric.get_metric('chebyshev') #uses chebyshev form to calculate distance
X=[[1,2],[3,4],[5,6]]
Y=[[7,8],[9,10]]
dist.pairwise(X,Y)
```

```python
from sklearn.metrics import DistanceMetric
dist=DistanceMetric.get_metric('minkowski') #uses minkowski form to calculate
X=[[1,2],[3,4],[5,6]]
Y=[[7,8],[9,10]]
dist.pairwise(X,Y)
```
